```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import mpl_toolkits
%matplotlib inline
```

```python
import plotly as py
import plotly.tools as tls
from plotly.offline import iplot
```

```python
import cufflinks as cf
```

```python
py.offline.init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100,3), columns = ['A', 'B', 'C'])
df.head()
df['A'] = df['A'].cumsum() + 20
df['B'] = df['B'].cumsum() + 20
df['C'] = df['C'].cumsum() + 20
```

```python
df.head()
```

#### iplot() is more interactive and shows a line chart's x, y values on hover

```python
df.iplot()
```

```python
plt.plot(df)
```

```python
df.plot()
```

```python
df.iplot(kind='box')
```

```python
df.iplot(x = 'A', y = 'B', mode = 'markers', size = 25)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.iplot(kind = 'bar', x = 'sex', y = 'survived', title = 'Survived', xTitle='Sex', yTitle='#Survived')
```

```python
titanic['sex'].value_counts()
```

```python
cf.getThemes()
```

```python
cf.set_config_file(theme='polar')
df.iplot(kind = 'bar', barmode='stack', bargap=0.5)
```

```python
df.iplot(kind = 'bar', barmode='stack', bargap=0.5)
```

```python
df.iplot(kind = 'barh', barmode='stack', bargap=0.5)
```

```python
df.iplot(kind = 'area')
```

```python
df[['A', 'B']].iplot(kind = 'spread')
```

```python
df3 = pd.DataFrame({'X': [10,20,30,20,10], 'Y': [10, 20, 30, 20, 10], 'Z': [10, 20, 30, 20, 10]})
df3.head()
```

```python
help(cf.datagen)
```

```python
cf.datagen.scatter3d(2, 150, mode = 'stocks').iplot(kind = 'scatter3d', x = 'x', y= 'y', z = 'z')
```

```python
df.iplot(kind='hist', bins = 25, barmode = 'overlay', bargap=0.5)
```

```python
cf.datagen.heatmap(20,20).iplot(kind = 'heatmap', colorscale='spectral', title='Cufflinks - Heatmap')
```
