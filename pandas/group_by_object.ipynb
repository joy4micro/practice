### The ideal thing we want to group by is that the column to have duplicate values. This means we want to create a segments of our dataset.

### agg(), it accepts a dictionary with key as the column and the value as the operation

```python
sectors.agg({'Profits': 'sum'})
```

```python
sectors.agg({'Profits': 'sum', 'Employees': 'mean'})
```

### agg(), accepts a list of operations

```python
sectors.agg(['sum', 'size', 'min'])
```

### iterate through groups

```python
for sector, data in sectors:
    print(data[['Company', 'Profits']])
```

```python
df = pd.DataFrame(columns=compaines.columns)
for sector, data in sectors:
    highest_profits_company_in_group = data.nlargest(2, 'Profits')
    df = df.append(highest_profits_company_in_group)
```

```python
df
```

```python
df = pd.DataFrame(columns=compaines.columns)
locationGroupBy = compaines.groupby('Location')
```

```python
locationGroupBy
```

```python
for city, data in locationGroupBy:
    highest_revenue_company_in_group = data.nlargest(1, 'Revenue')
    df = df.append(highest_revenue_company_in_group)
```
