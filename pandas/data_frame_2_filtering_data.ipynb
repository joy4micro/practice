#### duplicated() has keep attribute and it takes any of the 3 values

##### first : Mark duplicates as ``True`` except for the first occurrence.
##### last : Mark duplicates as ``True`` except for the last occurrence.
##### False : Mark all duplicates as ``True``.

```python
nameage.duplicated(keep=False)
```

```python
nameage.duplicated(keep='first')
```

```python
nameage.duplicated(keep='last')
```

##### There may be a situation when we want to find only the unique values, we can achieve this by using ~

```python
mask = ~nameage.duplicated(keep=False)
```

```python
mask
```

```python
nameage[mask]
```

#### We can use nunique() to get the unique length of a series or a dataframe. By default, the attribute dropna is set to True, so it doesn't include null values when it returns the length.

```python
# employees.Team.nunique() or
employees.Team.nunique(dropna=True)
```

```python
employees.Team.nunique(dropna=False)
```

```python
employees.nunique()
```

### unique() vs nunique() 
#### unique() returns all the unique values in a series, whereas, nunique() returns the length removing the null value.
