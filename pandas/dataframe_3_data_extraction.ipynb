#### Difference between loc and iloc
### For loc
##### If an index is a non numeric, e.g, Film. loc can be used to return the row of the non numeric index. Eg. movies.loc['Goldfinger']
##### If we pass any integer index say, movies.loc[0], it will throw an exception
### For Iloc
##### If an index is a numeric, even though we set an index to be non numeric, it will have number index managed internally. iloc can be used to return the row of the numeric index. Eg. movies.iloc[0]
##### If we pass any non integer index say, movies.iloc['Goldfinger'], it will throw an exception

##### Sometimes, we need to set value but the non integer index might have more than 1 records, in this case the values will be set in more than one record

##### So, we can use iloc instead of loc firstly by reseting the index and check for the value
##### movies[movies.Film == 'Casino Royale']

#### We need to use loc or iloc when we assign a value because when we use loc or iloc, it uses a reference to the actual dataframe and it will change the value
###### movies.loc[['Dr. No'], ['Actor', 'Budget', 'Bond Actor Salary']] = ['Sean Connery', 7000000, 6000000]
###### Where as, if we assign a value without loc or iloc, it won't refer to the original dataframe, it will give a copy of the actual dataframe and the value won't be changed
###### movies[['Dr. No'], ['Actor', 'Budget', 'Bond Actor Salary']] = ['Sean Connery', 7000000, 6000000]
