### fillna() to fill a value on null values.
#### fillna() on dataframe is better to use when all the datatypes of the columns are similar. If the columns datatypes are different, then we can use fillna() on series.

```python
nba.fillna(0)
```

#### Here, all the NaN values in salary is replaced with 0 which is good, but it has changed the valus to 0 for College also and it does not make any sense. So, we use fillna() on series.

```python
nba.Salary.fillna(value = 0, inplace=True)
```

```python
nba
```

```python
nba.College.fillna(value = 'No College', inplace=True)
```

```python
nba
```

### Rank values with .rank()

```python
nba = pd.read_csv("../../datasets/nba.csv").dropna(how='all')
```

```python
nba.Salary = nba.Salary.fillna(value = 0).astype('int')
```

```python
nba.College = nba.College.fillna(value = 'No College')
```

```python
nba
```

###### Suppose, we want to rank the salaries based on the income. Say, player with the highest salary has the rank of 1, the player with the second highest salary with the rank of 2 etc.

```python
nba['Salary_rank'] = nba.Salary.rank(ascending=False).astype('int')
```

```python
nba
```

```python
nba.sort_values(by='Salary', ascending=False)
```
