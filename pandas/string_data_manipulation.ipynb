### Split strings based on split()

```python
chicago = pd.read_csv('../../datasets/chicago.csv').dropna(how='all')
```

```python
chicago['Last Name'] = chicago.Name.str.split(",").str.get(0)
```

```python
chicago
```

```python
chicago['Last Name'].value_counts()
```

```python
chicago['first name'] = chicago.Name.str.split(",").str.get(1).str.split(" ").str.get(2)
```

```python
chicago.head()
```

```python
chicago['first name'].value_counts()
```

### The expand and n parameter of the str.split method

##### Returns a dataFrame rather than series of python list and values are separated across columns

```python
chicago[['First Name', 'Last Name']] = chicago['Name'].str.split(',' , expand=True)
```

```python
chicago
```

```python
chicago['Position Title'].str.split(' ' , expand=True)
```

### We can see many columns are having values as None because any of the column rows has a long value expanding to the column 8 in top table shown above
######  We can use n to split the values, e.g, n = 1 split the value by a total 1 space 
###### FRM OF MACHINISTS - AUTOMOTIVE would return a dataframe with 2 columns 
###### 1 is FRM and the 2nd is OF MACHINISTS - AUTOMOTIVE

```python
chicago['Position Title'].str.split(' ' , expand=True, n=1)
```

```python
chicago['Position Title'].str.split(' ' , expand=True, n=2)
```

```python
chicago[['First Title World', 'Remaining words']] = chicago['Position Title'].str.split(' ' , expand=True, n=1)
```
