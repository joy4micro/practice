#### Difference between loc and iloc
### For loc
##### If an index is a non numeric, e.g, Film. loc can be used to return the row of the non numeric index. Eg. movies.loc['Goldfinger']
##### If we pass any integer index say, movies.loc[0], it will throw an exception
### For Iloc
##### If an index is a numeric, even though we set an index to be non numeric, it will have number index managed internally. iloc can be used to return the row of the numeric index. Eg. movies.iloc[0]
##### If we pass any non integer index say, movies.iloc['Goldfinger'], it will throw an exception

#### We need to use loc or iloc when we assign a value because when we use loc or iloc, it uses a reference to the actual dataframe and it will change the value
###### movies.loc[['Dr. No'], ['Actor', 'Budget', 'Bond Actor Salary']] = ['Sean Connery', 7000000, 6000000]
###### Where as, if we assign a value without loc or iloc, it won't refer to the original dataframe, it will give a copy of the actual dataframe and the value won't be changed
###### movies[['Dr. No'], ['Actor', 'Budget', 'Bond Actor Salary']] = ['Sean Connery', 7000000, 6000000]

## Rename
#### rename one or more index/column label in a dataframe. It returns a brand new dataframe.

```python
movies = pd.read_csv('../datasets/jamesbond.csv', index_col='Film')
```

```python
movies.head()
```

```python
movies.rename(mapper = 
              {
                'Dr. No' : 'Doctor No',
                'Goldfinger' : 'Gold Finger'
              })
```

### Sometimes, we can explicitly specify the axis.
##### axis : {0 or 'index', 1 or 'columns'}, The default is 'index' or  0

### Note: 
##### mapper or index attribute will returns a new dataframe and it's immutable, so we can use inplace=True

```python
movies.rename(mapper = 
              {
                'Dr. No' : 'Doctor No',
                'Goldfinger' : 'Gold Finger'
              }, axis=0)
```

```python
movies.head()
```

```python
movies.rename(mapper = 
              {
                'Dr. No' : 'Doctor No',
                'Goldfinger' : 'Gold Finger'
              }, axis='index')
```

#### axis = 'rows' is not present, so it takes index as the axis value

```python
movies.rename(mapper = 
              {
                'Dr. No' : 'Doctor No',
                'Goldfinger' : 'Gold Finger'
              }, axis='rows')
```

### We can use index attribute instead of mapper. But, when we use index attribute, we cannot use axis attribute.
### Also, wecan use columns attribute for coulmn changes inetad of mapper and axis

```python
movies = pd.read_csv('../datasets/jamesbond.csv', index_col='Film')
```

```python
movies.rename(index = 
              {
                'Dr. No' : 'Doctor No',
                'Goldfinger' : 'Gold Finger'
              }, inplace=True)
```

```python
movies.head()
```

```python
movies = pd.read_csv('../datasets/jamesbond.csv', index_col='Film')
```

### Filter a dataframe with the query method
#### There are 2 limitations
##### 1) The arguments of the query method will be a string
##### 2) The column name passed to the query should not contain any space in between, we can always rename the spaces with _ 

##### The query method works faster than the access method according to the document

##### .query('column == "value"')

```python
movies.columns = ['Year', 'Actor', 'Director','Box_Office', 'Budget', 'Salary']
```

```python
movies.head(1)
```

```python
movies.query('Actor == "Sean Connery"')
```

```python
movies.query('Box_Office > 500')
```

```python
movies.query('Actor == "Sean Connery" and Box_Office > 500')
```

```python
movies.query('Actor in ["Sean Connery", "Daniel Craig"]')
```
